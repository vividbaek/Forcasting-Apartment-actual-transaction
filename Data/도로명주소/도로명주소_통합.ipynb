```python
#서울특별시 글자 제거 및 xlsx파일로 변경 (xls는 오래전 엑셀)

import pandas as pd
import os

# 기본 파일 경로
base_path = "C:\\Users\\vivid\\Desktop\\DAT\\부동산_프로젝트\\도로명 주소\\original\\"
base_filename = "도로명_조회_목록 "  # 띄어쓰기 포함

# 파일 전체 반복문
for i in range(25):
    if i == 0:
        filename = f"{base_filename}.xls"
    else:
        filename = f"{base_filename}({i}).xls"
    
    # 전체 파일 경로
    full_path = os.path.join(base_path, filename)
    new_full_path = os.path.join(base_path, filename.replace('.xls', '.xlsx'))  # 새 파일 경로

    # 파일 읽기
    try:
        data = pd.read_excel(full_path)
        # 첫 번째 열 이름 변경
        first_column = data.columns[0]
        new_column_name = first_column[6:]  # 첫 6자리 제거
        data = data.rename(columns={first_column: new_column_name})

        # 변경된 데이터를 .xlsx 형식으로 저장
        data.to_excel(new_full_path, index=False, engine='openpyxl')
        print(f"Updated {filename} to {new_full_path}")
    except Exception as e:
        print(f"Error processing {filename}: {e}")
```

```python
import pandas as pd
import numpy as np
import os

#각 구별로 도로명 리스트를 저장할 딕셔너리
district_roads = {}

#파일 전체 반복문
for i in range(25):
    if i == 0:
        filename = f"{base_filename}.xlsx"
    else:
        filename = f"{base_filename}({i}).xlsx"
    
    #전체 파일 경로
    full_path = os.path.join(base_path, filename)

    #파일 읽기
    try:
        data = pd.read_excel(full_path)
        district_name = data.columns[0]
        road_names = data['Unnamed: 1'][1:].tolist()
        #딕셔너리 저장
        district_roads[district_name] = road_names

    except Exception as e:
        print(f"Error reading {filename} : {e}")

#각 지역별로 리스트가 나오는 것은 확인할 수 있음!
for district, roads in district_roads.items():
    print(f"{district} = {roads}")
```

```python
#csv파일로 저장하기

district_data = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in district_roads.items()]))

csv_path = os.path.join(base_path, 'district_roads.csv')
district_data.to_csv(csv_path, index=False, encoding="utf-8-sig")

print(f"완료")
```
